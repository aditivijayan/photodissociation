```python
import seaborn as sns
import matplotlib as mpl
import _init_
from constants import *
from set_path import *
from config import *
from functions import *
import scipy.integrate as integrate
```

```python
cbarx = 0.141
cbheight = 0.02
cbary = 0.9
cblen = 0.35
dx1 = 0.4
```

```python
f_diss = 0.1
X = 0.74 
F0 = 2.1e7
G0 = 1 * 2. * f_diss * F0 * mp/X
r0 = 250. * pc
v0 = 170. * kmps
t0 = r0/v0
Sigma0        = 52.*Msun_pcsq
sat_time = Sigma0/G0
xi = np.log(G0 *t0/Sigma0)
np.log(sat_time/(t0)), sat_time/Myr, t0/Myr, xi
```

```python
rmax = 10.*kpc
v0   = 170. * kmps
arr  = np.linspace(1.,int(rmax/r0),500)
da   = arr[1] - arr[0]
a_scale = r0*arr

Gamma = 0.5
xcrit = np.log(Gamma)
xhigh = 0.99*xcrit
xlow  = 1.5*xcrit
xval  = np.linspace(xlow, xhigh, 10)
```

# Acc law and tdyn for Fixed Area Case

```python
f, ax = plt.subplots(3, 1, gridspec_kw = {'wspace':0.05, 'hspace':0.17},figsize=(8,20))
colors =  sns.color_palette("hls", 8)


ls = ['-', '--', '--']
i=j = 0

t     = np.zeros((xval.shape[0],arr.shape[0]))
cmap  = plt.get_cmap('jet',xval.shape[0])
frac_diss = np.zeros((xval.shape[0],arr.shape[0]))

for x in xval:
    i=0
    for a in arr:
        t[j,i] = integrate.quad(da_Uax_fAr, 1, a, args=(Gamma, x, r0/v0))[0] 
        
        frac_diss[j,i]      = fHI_anly_fAr(a, gamma, x, t0, G0, Sigma0)
        i+=1
    
    t[j,0] = 0.0
     
    U_ax = Uax_fAr(arr, Gamma, x) *v0/kmps
    ax[0].plot(a_scale/kpc, U_ax, lw=3.0, color=cmap(j))

    frac_diss[frac_diss>1.] = 1.
    
    ax[1].plot(a_scale/kpc, t[j]/Myr, lw=3.0, color=cmap(j))
    ax[2].plot(a_scale/kpc, 1.-frac_diss[j], lw=3.0, color=cmap(j))

    j+=1
    
ax[0].set_ylabel(r'U$_a (x)$ [km s$^{-1}$]')
ax[0].set_title(r"$v_0=$%d km s$^{-1}$"%(v0/kmps) + '\n' +  r"$r_0=$%d pc"%(r0/pc))

ax[1].tick_params(axis='y', which='both', labelleft=False, labelright=True, right=True)
ax[1].yaxis.set_label_position("left")
ax[1].set_ylabel(r'$t_{\rm dyn}=\int \frac{dr}{Uax}$ [Myr]', labelpad=30)
ax[1].yaxis.tick_left()


# ax[2].set_ylim(0.0, 0.5)
# ax[2].set_yscale('log')
ax[2].set_ylabel(r'$\mathrm{f}_{\rm H2}$', labelpad=20)
# ax[2].set_yscale('log')

plt.setp(ax[0], 'xlabel', 'r [kpc]')
plt.setp(ax[1:], 'xlabel', 'r [kpc]')
plt.setp(ax[1:], 'xlim',(r0/kpc, rmax/kpc))
ax[2].set_xlim(0.0, 10.)
ax[2].set_ylim(0., 1.1)

norm = mpl.colors.Normalize(vmin=xlow,vmax=xhigh)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
plt.colorbar(sm, ticks=xval,pad=0.01, location='right')
file_name = os.path.join(fig_path, 'hi_frac_fAR.jpeg')
# plt.savefig(file_name, bbox_inches='tight', dpi=160 )
```
